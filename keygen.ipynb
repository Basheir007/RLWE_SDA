```python
import numpy as np
import hashlib
import time
import matplotlib.pyplot as plt


def hashMd(x):
    t=str(x)
    result = hashlib.md5(t.encode()) 
    # printing the equivalent hexadecimal value. 
    print("The hexadecimal hash of PlainText is equivalent: ", end ="") 
    print(result.hexdigest()) 
    
def checkValidity(p,q):
    if hashMd(p) in hashMd(q):
        print("VALID DATA")
    else:
        print("modified")

def gen(x):
    return x.random.randint(1,5, size=(3, 3))

key_gen=gen(np)

public_key =key_gen[:,0]
private_key = np.reciprocal(public_key.astype(np.float32))


#error distribution
def error():
    err=np.random.uniform(1,3, size=(3))
    return err
err=error()
err=err.astype(int)
print("error",err)


#encryption 

def encryption(x):
    cipher=x*public_key+err
    return cipher


#decryption

def decryption(x):
    decipher= x*private_key-err
    return decipher.round()

    

print(key_gen)

print("public_key",key_gen[:,0])

start = time.time()
print("private_Key",private_key)
smart = np.loadtxt('mytext.txt')
smart = smart.reshape((188,3))
smart=smart.round()
#np.savetxt('mytext.txt',smart,fmt='%-7.2f')

print("\n======plainText======\n",smart)
print("\n===hashOfplainTextI======\n")
hashMd(smart)

print("size of PlainText=",smart.shape)

cipher=encryption(smart)


print("\n======cipherText======\n",cipher)
print("size of cipher=",cipher.shape)
dicipher=decryption(cipher)

print("\n======decipherText======\n",dicipher)
print("size of decipher=",dicipher.shape)
end = time.time()
totalTime=end-start
print("total time taken====((((",round(totalTime,3),"))))")
```
